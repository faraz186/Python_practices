##### 3. Find and display the largest number of a list without using built-in function max(). Your program should ask the user to input values in list from keyboard. 

```python

lst = input("Enter numbers separated by space: ").split()

numbers = [int(x) for x in lst]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest number in the list is:", largest)
```

##### 4. Write a program that rotates the element of a list so that the element at the first index moves to the second index, the element in the second index moves to the third index, etc., and the element in the last index moves to the first index. 

```python
list = input("Enter elements separated by space: ").split()

# Rotate the list
rotated_list = [list[-1]] + list[:-1]

# Display result
print("List after rotation:")
print(rotated_list)
```

##### 6. Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It should print the date in the form March 12, 2021.  

```python
date = input("Enter date (mm/dd/yyyy): ")

month, day, year = date.split('/')

month = int(month)

months = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

print(months[month - 1], day + ",", year)
```

##### 7. Write a program with a function that accepts a string from keyboard and create a new string after converting character of each word capitalized. For instance, if the sentence is "stop and smell the roses." the output should be "Stop And Smell The Roses" 

```python
def capitalize_words(sentence):
    words = sentence.split()
    new_sentence = ""

    for word in words:
        new_sentence += word[0].upper() + word[1:] + " "

    return new_sentence.strip()


# Input from user
text = input("Enter a sentence: ")

# Call the function
result = capitalize_words(text)

# Display result
print("Capitalized sentence:")
print(result)
```

##### 8. Find the sum of each row of matrix of size m x n. For example for the following matrix output will be 
like this : 
Sum of row 1 = 32 
Sum of row 2 = 31 
Sum of row 3 = 63 

```python
# Input number of rows and columns
m = int(input("Enter number of rows: "))
n = int(input("Enter number of columns: "))

matrix = []

# Input matrix elements
print("Enter matrix elements row-wise:")
for i in range(m):
    row = []
    for j in range(n):
        row.append(int(input(f"Element [{i}][{j}]: ")))
    matrix.append(row)

# Calculate and display sum of each row
for i in range(m):
    row_sum = 0
    for j in range(n):
        row_sum += matrix[i][j]
    print(f"Sum of row {i+1} = {row_sum}")
```

##### 9. Write a program to add two matrices of size n x m.

```python
# Input number of rows and columns
n = int(input("Enter number of rows: "))
m = int(input("Enter number of columns: "))

matrix1 = []
matrix2 = []

# Input first matrix
print("Enter elements of first matrix:")
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input(f"Matrix1[{i}][{j}]: ")))
    matrix1.append(row)

# Input second matrix
print("Enter elements of second matrix:")
for i in range(n):
    row = []
    for j in range(m):
        row.append(int(input(f"Matrix2[{i}][{j}]: ")))
    matrix2.append(row)

# Add matrices
result = []
for i in range(n):
    row = []
    for j in range(m):
        row.append(matrix1[i][j] + matrix2[i][j])
    result.append(row)

# Display result
print("Sum of the two matrices:")
for row in result:
    print(row)
```

##### 10. Write a program to multiply two matrices 

```python
# Input dimensions of first matrix
r1 = int(input("Enter number of rows of first matrix: "))
c1 = int(input("Enter number of columns of first matrix: "))

# Input dimensions of second matrix
r2 = int(input("Enter number of rows of second matrix: "))
c2 = int(input("Enter number of columns of second matrix: "))

# Check if multiplication is possible
if c1 != r2:
    print("Matrix multiplication not possible.")
else:
    matrix1 = []
    matrix2 = []

    # Input first matrix
    print("Enter elements of first matrix:")
    for i in range(r1):
        row = []
        for j in range(c1):
            row.append(int(input(f"Matrix1[{i}][{j}]: ")))
        matrix1.append(row)

    # Input second matrix
    print("Enter elements of second matrix:")
    for i in range(r2):
        row = []
        for j in range(c2):
            row.append(int(input(f"Matrix2[{i}][{j}]: ")))
        matrix2.append(row)

    # Initialize result matrix with zeros
    result = []
    for i in range(r1):
        result.append([0] * c2)

    # Multiply matrices
    for i in range(r1):
        for j in range(c2):
            for k in range(c1):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    # Display result
    print("Resultant matrix after multiplication:")
    for row in result:
        print(row)
```
